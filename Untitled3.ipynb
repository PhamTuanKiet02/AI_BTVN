```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctie

wind_speed = ctie.Antecedent(np.arange(0.6,3.1,1),'wind_speed')
temperature = ctie.Antecedent(np.arange(20.3,40.2,1),'temperature')

rainfall = ctie.Consequent(np.arange(0.3,9,1),'rainfall')

wind_speed['very low WS'] = fuzz.trimf(wind_speed.universe,[0.5,0.5,1])
wind_speed['low WS'] = fuzz.trimf(wind_speed.universe,[0.5,1.2,2.2])
wind_speed['normal WS'] = fuzz.trimf(wind_speed.universe,[2,2.4,2.4])
wind_speed['high WS'] = fuzz.trimf(wind_speed.universe,[2.2,3,4])
wind_speed['very high WS'] = fuzz.trimf(wind_speed.universe,[2.2,2.8,3])

temperature['very low TP'] = fuzz.gaussmf(temperature.universe,20.3,5)
temperature['low TP'] = fuzz.gaussmf(temperature.universe,23,5)
temperature['normal TP'] = fuzz.gaussmf(temperature.universe,27.4,5)
temperature['high TP'] = fuzz.gaussmf(temperature.universe,34.4,5)
temperature['very high TP'] = fuzz.gaussmf(temperature.universe,41,5)

rainfall['very low RF'] = fuzz.trimf(rainfall.universe,[0.2,0.2,1.5])
rainfall['low RF'] = fuzz.trimf(rainfall.universe,[0.2,1.2,3.2])
rainfall['normal RF'] = fuzz.trimf(rainfall.universe,[1.5,3.6,4])
rainfall['high RF'] = fuzz.trimf(rainfall.universe,[3.2,4.2,5.5])
rainfall['very high RF'] = fuzz.trimf(rainfall.universe,[3.3,3.6,4])

wind_speed.view()
temperature.view()
rainfall.view()

rule1 = ctie.Rule(wind_speed['very high WS']&temperature['normal TP'],rainfall['high RF'])
rule2 = ctie.Rule(wind_speed['very low WS']&temperature['very low TP'],rainfall['very low RF'])
rule3 = ctie.Rule(wind_speed['very high WS']&temperature['very high TP'],rainfall['very high RF'])
rule4 = ctie.Rule(wind_speed['low WS']&temperature['low TP'],rainfall['very low RF'])
rule5 = ctie.Rule(wind_speed['high WS']&temperature['very high TP'],rainfall['high RF'])

rainfall_ctie= ctie.ControlSystem([rule1,rule2,rule3,rule4,rule5])
rainfalling = ctie.ControlSystemSimulation(rainfall_ctie)

rainfalling.input['wind_speed']=2.5
rainfalling.input['temperature']=26
rainfalling.compute()
print(rainfalling.output['rainfall'])
rainfall.view(sim=rainfalling)
```

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctie

taste = ctie.Antecedent(np.arange(0,10,1),'taste')
environment = ctie.Antecedent(np.arange(0,11,1),'environment')
budget = ctie.Antecedent(np.arange(0,11,1),'budget')
location = ctie.Antecedent(np.arange(0,4,1),'location')
user_review = ctie.Antecedent(np.arange(0,5,1),'user_review')

preference = ctie.Consequent(np.arange(0,15,1),'preference')

taste['low'] = fuzz.gaussmf(taste.universe,0,4)
taste['medium'] = fuzz.gaussmf(taste.universe,3,5)
taste['high'] = fuzz.gaussmf(taste.universe,10,3)

environment['low'] = fuzz.gaussmf(environment.universe,0,4)
environment['medium'] = fuzz.gaussmf(environment.universe,5,5)
environment['high'] = fuzz.gaussmf(environment.universe,10,4)

budget['Less Preferred'] = fuzz.gaussmf(budget.universe,6,5)
budget['Preferred'] = fuzz.gaussmf(budget.universe,4,6)
budget['More Preferred'] = fuzz.gaussmf(budget.universe,0,5)

user_review['Less Preferred'] = fuzz.gaussmf(user_review.universe,4,3)
user_review['Preferred'] = fuzz.gaussmf(user_review.universe,4,5)
user_review['More Preferred'] = fuzz.gaussmf(user_review.universe,0,5)

location['Close'] = fuzz.gaussmf(location.universe,0,2)
location['Near'] = fuzz.gaussmf(location.universe,2,2)
location['Far'] = fuzz.gaussmf(location.universe,3,2)

preference['lvl 1'] = fuzz.trimf(preference.universe,[0,0,3])
preference['lvl 2'] = fuzz.trimf(preference.universe,[0,5,7])
preference['lvl 3'] = fuzz.trimf(preference.universe,[4,5,5])
preference['lvl 4'] = fuzz.trimf(preference.universe,[5,10,13])
preference['lvl 5'] = fuzz.trimf(preference.universe,[8,13,15])

taste.view()
environment.view()
budget.view()
user_review.view()
location.view()
preference.view()

rule1 = ctie.Rule(taste['low']&environment['low']&budget['More Preferred']&location['Far']&user_review['More Preferred'],preference['lvl 1'])
rule2 = ctie.Rule(taste['low']&environment['medium']&budget['Preferred']&location['Far']&user_review['More Preferred'],preference['lvl 2'])
rule3 = ctie.Rule(taste['medium']&environment['medium']&budget['More Preferred']&location['Far']&user_review['More Preferred'],preference['lvl 3'])
rule4 = ctie.Rule(taste['high']&environment['medium']&budget['Preferred']&location['Far']&user_review['Less Preferred'],preference['lvl 4'])
rule5 = ctie.Rule(taste['high']&environment['high']&budget['Less Preferred']&location['Close']&user_review['Less Preferred'],preference['lvl 5'])

preference_ctie= ctie.ControlSystem([rule1,rule2,rule3,rule4,rule5])
preferencing = ctie.ControlSystemSimulation(preference_ctie)

preferencing.input['taste']=2
preferencing.input['environment']=1
preferencing.input['budget']=2
preferencing.input['location']=3
preferencing.input['user_review']=2
preferencing.compute()
print(preferencing.output['preference'])
preference.view(sim=preferencing)
```
